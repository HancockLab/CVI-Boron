```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_table('leaf_B_Atwell.csv',sep=',')
df
```

```python
import matplotlib.ticker as ticker
cm = 1/2.54
fig = plt.figure(figsize=(8*cm,6*cm))
ax=sns.histplot(x='phenotype_value',
             data=df,
             stat='density',
             color='gray',
             bins=30)
# ax.axvline(x=df[df['accession_name'] == 'Cvi-0'].phenotype_value.values,linewidth=1,color='0',alpha=0.6,linestyle='--') 
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.set_xlabel('Leaf [B] (µg/g DW)')
ax.set_xlim(60,120)
sns.despine(offset=10, trim=True)
plt.tight_layout()
# plt.savefig("Fig1a.png", format="png",bbox_inches="tight",facecolor="w",dpi=400)
plt.savefig("Fig1a.pdf", format="pdf")
```
