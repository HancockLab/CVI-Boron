```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import matplotlib.patches as mpl_patches
from scipy.stats import mannwhitneyu
import matplotlib.ticker as ticker
```

```python
df=pd.read_table('DH_leaf_B_wGenotype.txt',header=None)
df=df.rename(columns={0:"Seq_ID", 1:"B", 2:"BOR1", 3:"BOR2"})
df
```

```python
df=df.loc[df['BOR1'] != '.']
df
```

```python
boxprops = {'edgecolor': 'black', 'linewidth': 0.5, 'facecolor': 'grey'}
lineprops = {'color': 'k', 'linewidth': 0.5}
boxplot_kwargs = dict({'boxprops': boxprops, 'medianprops': lineprops,
                       'whiskerprops': lineprops, 'capprops': lineprops,
                       'width': 0.6})
@ticker.FuncFormatter
def major_formatter(x, pos):
    y = x /1000000
    return "%.0f" % y
cm = 1/2.54
plt.figure(figsize=(5*cm,5*cm))
ax=sns.boxplot(x='BOR1', 
            y='B',
            color='grey',
            data=df,
            linewidth=1.5,
            fliersize=0,
            **boxplot_kwargs)
ax=sns.swarmplot(x='BOR1', 
            y='B',
            color='0',
            data=df,
            s=2.5)
sns.despine(trim=True,offset=10)
plt.tick_params(labelsize=8)
plt.savefig("Fig4d.pdf", format="pdf",bbox_inches="tight",facecolor="w")
```

```python
boxprops = {'edgecolor': 'black', 'linewidth': 0.5, 'facecolor': 'grey'}
lineprops = {'color': 'k', 'linewidth': 0.5}
boxplot_kwargs = dict({'boxprops': boxprops, 'medianprops': lineprops,
                       'whiskerprops': lineprops, 'capprops': lineprops,
                       'width': 0.6})
@ticker.FuncFormatter
def major_formatter(x, pos):
    y = x /1000000
    return "%.0f" % y
cm = 1/2.54
plt.figure(figsize=(8*cm,6*cm))
ax=sns.boxplot(x='BOR1', 
            y='B',
            color='grey',
            data=df,
            linewidth=1.5,
            fliersize=0,
            **boxplot_kwargs)
ax=sns.swarmplot(x='BOR1', 
            y='B',
            color='0',
            hue='BOR2',
            palette=["0", "red"],
            data=df,
            s=2.8)
sns.despine(trim=True,offset=10)
plt.tick_params(labelsize=8)
# plt.savefig("Fig4d.pdf", format="pdf",bbox_inches="tight",facecolor="w")
```
